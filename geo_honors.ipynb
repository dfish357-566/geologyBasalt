```python
import pandas as pd
import matplotlib.pyplot as plt

S1 = pd.read_csv('LRMCC_S1_15058_248.csv')  
S2 = pd.read_csv('LRMCC_S2_15555_971.csv')  
S3 = pd.read_csv('LRMCC_S3_70017_541.csv')  
S4 = pd.read_csv('LRMCC_S4_70035_193.csv')
```

```python
def prepareDF(df):
    # remove first column: "Unnamed: 0"
    columns = df.columns[1:]
    print(columns)

    # convert all values to floats, replace empty spaces with 0
    for column in columns:
      df[column] = df[column].map(lambda x: ''.join([i for i in x if i.isdigit() or i=="."]))
      df = df.replace(r'^\s*$', 0, regex=True)
      df[column] = df[column].astype(float)

    df = df.set_index('Unnamed: 0')
    df = df.drop("Total")
    df = df.T
    print(df.to_string())
    return df
    #ax = df.plot.bar(rot=0)
```

```python
S1 = prepareDF(S1)
S2 = prepareDF(S2)
S3 = prepareDF(S3)
S4 = prepareDF(S4)
```

```python
# bar plot of aluminum allocation
plt.title("Sample 1: 15058_248 (Low titanium)")
S1["Al2O3"].plot(kind='bar', width=1, figsize=(15,10))
```

```python
#S1["FeO"].mean()/len(S1.index)*10

def getPercentages(dfwa):
  df = dfwa.iloc[: , :-1]
  returnFrame = pd.DataFrame()
  cols = df.columns
  for col in cols:
    returnFrame[col] = [df[col].mean()/len(cols)*10]
  return returnFrame
```

```python
percentagesS1 = getPercentages(S1)
percentagesS2 = getPercentages(S2)
percentagesS3 = getPercentages(S3)
percentagesS4 = getPercentages(S4)


print('S1_15058_248', '\n', percentagesS1.to_string(), percentagesS1.sum(axis=1))
print('S2_15555_971', '\n', percentagesS2.to_string(), percentagesS2.sum(axis=1))
print('S3_70017_541', '\n', percentagesS3.to_string(), percentagesS3.sum(axis=1))
print('S4_70035_193', '\n', percentagesS4.to_string(), percentagesS4.sum(axis=1))
```
